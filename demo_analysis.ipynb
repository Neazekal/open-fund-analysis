```python
def main():
    # Chỉ định folder chứa CSV quỹ
    folder = r"data/stock_fund/*.csv"

    # Lấy danh sách file CSV quỹ
    import glob
    fund_files = sorted(glob.glob(folder))
    print(f"Tìm thấy {len(fund_files)} file quỹ")

    # So sánh và tính toán chỉ số
    results = compare_funds(fund_files, rf_annual=0.03)

    if results.empty:
        print("Không có dữ liệu để phân tích (chưa đủ 1 năm).")
        return

    # In bảng kết quả
    print("\n=== Bảng kết quả ===")
    print(results.round(4))

    # Vẽ một số bar chart
    print("\nVẽ biểu đồ...")
    bar_rank(results, "sharpe", "Sharpe Ratio")
    bar_rank(results, "cagr", "CAGR")
    bar_rank(results, "vol_annual", "Annualized Volatility")
    bar_rank(results, "ret_1y", "1-Year Return")
```

```python
if __name__ == "__main__":
    main()
```
